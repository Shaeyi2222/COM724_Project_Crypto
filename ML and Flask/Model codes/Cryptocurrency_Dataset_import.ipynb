```python
import yfinance as yf
import pandas as pd
import zipfile
import os

def get_top_100_crypto_data(start_date, end_date):
    # Get the top 100 cryptocurrencies
    crypto_symbols = [
    "ETH-USD", "BNB-USD", "DOGE-USD", "SOL-USD", "TRX-USD",
    "MATIC-USD", "SHIB-USD", "LTC-USD", "WBTC-USD",
    "BCH-USD", "AVAX-USD", "XRP-USD", "ADA-USD",
    "LINK-USD", "HBAR-USD", "ARB-USD", "ETC-USD",
    "XMR-USD", "APT-USD", "BTC-USD"
]


    # Download data for each cryptocurrency
    data = yf.download(crypto_symbols, start=start_date, end=end_date, interval='60m', group_by='ticker')

    # Create a dictionary to store data for each cryptocurrency
    crypto_data = {}
    for symbol in crypto_symbols:
        df = data[symbol].copy()
        df['Symbol'] = symbol
        crypto_data[symbol] = df

    return crypto_data

def save_to_zip(crypto_data):
    # Save each cryptocurrency data to individual CSV files
    for symbol, df in crypto_data.items():
        filename = f'{symbol}.csv'
        df.to_csv(filename)

    # Zip all the CSV files
    with zipfile.ZipFile('crypto_data.zip', 'w') as zip_file:
        for symbol in crypto_data.keys():
            zip_file.write(f'{symbol}.csv')

    # Remove the individual CSV files after zipping
    for symbol in crypto_data.keys():
        filename = f'{symbol}.csv'
        os.remove(filename)

if __name__ == "__main__":
    # Define the date range for historical data
    start_dates = ['2022-08-14']
    end_dates = ['2023-08-14']

    for start_date, end_date in zip(start_dates, end_dates):

        # Get data for the top 100 cryptocurrencies
        crypto_data = get_top_100_crypto_data(start_date, end_date)

        # Save the data to a zipped CSV file
        save_to_zip(crypto_data)
```

```python
if __name__ == "__main__":
    # Define the date range for historical data
    start_date = '2021-12-01'
    end_date = '2022-07-19'

    # Get data for the top 100 cryptocurrencies
    crypto_data_2 = get_top_100_crypto_data(start_date, end_date)

    # Save the data to a zipped CSV file
    save_to_zip(crypto_data_2)
```
